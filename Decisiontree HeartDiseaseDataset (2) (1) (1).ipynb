```python
#Hyperparameter Tunnings	           Description	                                 Effect on Performance
#max_depth:-	 Maximum depth of the tree	-  Controls overfitting. Shallow = underfit, deep = overfit
#min_samples_split:-	Minimum number of samples required to split a node	- Higher value → more conservative tree (less complex)
#min_samples_leaf:-	Minimum number of samples required in a leaf node -	Prevents creation of small leaf nodes; reduces overfitting
#max_leaf_nodes:-	Maximum number of leaf nodes	Limits complexity by bounding the total number of final decisions (leaves)
#max_features:-	Number of features to consider when splitting a node	Introduces randomness → useful in ensembles (e.g., Random Forest)
```

```python
#1. Label Encoding
#Definition:- Converts categories into numeric labels (e.g., "red" → 0, "blue" → 1)
#Output Format:-	Single column with integer values
#Use Case:-	For ordinal data (where order matters), or tree-based models (like Decision Trees, Random Forests)
#Example:-	["cat", "dog", "mouse"] → [0, 1, 2]
#Problem:-	Can mislead linear models into thinking one category is greater than another (e.g., "dog" > "cat")

#2. One-Hot Encoding
#Definition:-	Converts each category into a new binary column
#Output Format;-	Multiple columns, each for a category
#Use Case:-	For nominal data (no order), and models that are sensitive to numerical magnitude (e.g., Logistic Regression, SVM, Neural Nets)
#Example:-	["cat", "dog", "mouse"] → [[1,0,0], [0,1,0], [0,0,1]]
#Problem:-	Increases dimensionality — especially with high-cardinality categorical variables


```
