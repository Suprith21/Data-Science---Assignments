```python
#1.prime number or not
num = int(input("Enter a number: "))
if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print(num, "is not a prime number.")
            break
    else:
        print(num, "is a prime number.")       
else:
    print(num, "is not a prime number.")
```

```python
#2.Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. 
#The program should then check if the user's answer is correct and display an appropriate message.
import random
n1 = random.randint(1, 10)
n2 = random.randint(1, 10)
user_ans = int(input(f"What is {n1} × {n2}? "))
crt_ans = n1*n2

if user_ans == crt_ans:
    print("Correct answer..")
else:
    print(f" Wrong,The correct answer is {crt_ans}.")
```

```python
# 4.write a program to count the number of words in a given text.
input_text="This is a sample text. This text will be used to demonstrate the word counter."
for ch in ".,":
    input_text=input_text.replace(ch, "")

input_text=input_text.lower()
print(input_text)
words=input_text.split()
word_count={}

for word in words:
    if word in word_count:
        word_count[word]+=1
    else:
        word_count[word]=1


for word, count in word_count.items():
    print(f"'{word}':{count}")
```

```python
#5.check palindrome or not
def clean_text(text):
    # Keep only alphanumeric characters and convert to lowercase
    return ''.join(char.lower() for char in text if char.isalnum())

def is_palindrome(s):
    if len(s)<=1:
        return True
    if s[0]==s[-1]:
        return is_palindrome(s[1:-1])
    else:
        return False

user_input=input("Enter a word or phrase: ")
cleaned_input=clean_text(user_input)

# Call the recursive function and print result
if is_palindrome(cleaned_input):
    print("It's a palindrome!")
else:
    print("It's not a palindrome.")
```
